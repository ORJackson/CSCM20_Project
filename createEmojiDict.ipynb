```python
import pandas as pd
from textacy.preprocessing import normalize
```

```python
emojisDf = pd.read_csv("emojis.csv", sep="&", lineterminator="\r",)
emojisDf.shape
emojisDf.head(20)
```

```python
emojisDf['emoji and translation'] = emojisDf['emoji and translation'].apply(lambda x: normalize.whitespace(x))

def getEmoji(emojiString):
    firstLetterIndex = emojiString.find(next(filter(str.isalpha, emojiString)))
    return normalize.whitespace(emojiString[0:firstLetterIndex].replace("\u200d", ""))

emojisDf['emoji'] = emojisDf['emoji and translation'].apply(lambda x: getEmoji(x))

def getTranslation(emojiString):
    firstLetterIndex = emojiString.find(next(filter(str.isalpha, emojiString)))
    return normalize.whitespace(emojiString[firstLetterIndex:].lower())

emojisDf['translation'] =  emojisDf['emoji and translation'].apply(lambda x: getTranslation(x))


emojisDf.drop(columns=["emoji and translation"],  inplace=True)

emojisDf.head()
```

```python
def numEmojisIs1(s):
    if len(s) == 1:
        return True
    else:
        return False

emojisDf['emoji1'] = emojisDf['emoji'].apply(lambda x : numEmojisIs1(x))
emojisDf.tail(100)
```

```python
emojisDf = emojisDf.loc[emojisDf["emoji1"] == True]
emojisDf.head()
emojisDf.drop(columns=["emoji1"],  inplace=True)
emojisDf.shape
```

```python

emojiDict = emojisDf.to_dict(orient='split')
emojiData = emojiDict["data"]

d = dict((x, y) for x, y in emojiData)

print(d)
```
